```python
import ultralytics
import cv2
import torch
```

### Processar dados

```python
# Baixa o modelo
model = ultralytics.YOLO("yolo11n.pt")

# Jogao o modelo para o dispositivo de computação
device =  "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

# processa o vídeo
results = model.track(source="video.mov", conf=0.1, iou=0.7)
```

### Salvar Vídeo

```python
# validar resultados
for r in results:
    print(r)
```

```python
# Cria o escritor do vídeo de output
# Você pode pesquisar para salvar em outros formatos
writer  = cv2.VideoWriter("output.avi", cv2.VideoWriter_fourcc(*"XVID"), 20, (2048, 1080))

# Escreve o vídeo de output
for r in results:
    writer.write(r.plot())
```
